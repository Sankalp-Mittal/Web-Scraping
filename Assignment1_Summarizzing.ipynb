#### TASK 3: ModiJi Crawler

1. Navigate to Narendra Modi's Wikipedia, and fool around. I cannot legally ask you to do anything else.
2. Write a script that _crawls_ the page, that is, it returns a list of all links on the page. Then for each link, it returns all links and so on. Do this for two levels.
**BONUS** : Write a script that can crawl for N levels, where N is user specified. 

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Narendra_Modi"
parser = 'html.parser'
```

```python
def crawl(url_in, N, n):
    if n > 0:
        resp = requests.get(url_in)
        soup = BeautifulSoup(resp.content, parser)
        links = []
        i = 1

        for link in soup.find_all('a', href=True):
            if link['href'] != "":
                if link['href'][:6] == "/wiki/":
                    link["href"] = "https://en.wikipedia.org" + link["href"]
                    # links.append(link['href'])
                elif link['href'][:9] == "//en.wiki":
                    link["href"] = "https:" + link["href"]
                    # links.append(link['href'])
                if "https://" in link["href"]:
                    if "wikipedia" in link["href"]:
                        links.append(link['href'])

        for link in links:
            for _ in range(N - n):
                print("   ", end="")
            print(f"{i}/{len(links)}) {link}")
            i += 1
            crawl(link, N, n - 1)
    else:
        return
```

```python
def crawl_driver(url_in, n):
    crawl(url_in, n, n)
```

```python
iterations = int(input("Enter number of levels to be crawled? "))
```

```python
crawl_driver(url, iterations)
```
